```python
#personal preference
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.float_format', lambda x: f'{x:.3f}')
```

```python
# Load the Excel file into a DataFrame
file_path = r'C:\Users\brian\OneDrive\Documents\Data Analytics\Data Sets\OnlineBankingData\CACU 2020ToPres.xlsx'
df = pd.read_excel(file_path)

# Ensure 'Description' and 'Memo' are treated as strings
df['Description'] = df['Description'].astype(str)
df['Memo'] = df['Memo'].astype(str)

# Concatenate 'Description' and 'Memo' into a new column 'descriptionAndMemo'
df["descriptionAndMemo"] = df["Description"] + " " + df["Memo"]

# Filter out transactions that have "point of sale" in 'descriptionAndMemo'
pos_transactions = df[~df['descriptionAndMemo'].str.contains("ACH Payment", case=False)]

# Remove POS transactions from the main DataFrame
df = df[df['descriptionAndMemo'].str.contains("ACH Payment", case=False)]

# Continue with grouping and analysis on the filtered DataFrame
grouped = df.groupby('descriptionAndMemo')

# Filter groups to find recurring transactions (appearing more than once)
recurring_transactions = pd.DataFrame(grouped.filter(lambda x: len(x) > 1))

# For recurring POS transactions, group and filter similarly
grouped_pos = pos_transactions.groupby('descriptionAndMemo')
recurring_pos_transactions = pd.DataFrame(grouped_pos.filter(lambda x: len(x) > 1))

# If you want to see a summary of recurring transactions (e.g., total amount, count)
summary_df = recurring_transactions.groupby('descriptionAndMemo')['Amount Debit'].agg(['sum', 'count'])

# And a summary for recurring POS transactions
summary_pos_df = recurring_pos_transactions.groupby('descriptionAndMemo')['Amount Debit'].agg(['sum', 'count'])

# Sort the summary dataframes by the 'count' column in descending order
summary_df = summary_df.sort_values(by='count', ascending=False)
summary_pos_df = summary_pos_df.sort_values(by='count', ascending=False)
```
